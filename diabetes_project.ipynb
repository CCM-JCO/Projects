```python
"""
Let's check if these features are normally distributed.
I'll use the Shapiro-Wilk Test to evaluate normality.
"""
from scipy.stats import shapiro

# Normality test
stat1, p1 = shapiro(df.BMI)
stat2, p2 = shapiro(df.BloodPressure)
stat3, p3 = shapiro(df.Glucose)

# Check
def response(p_value):
    alpha = 0.05
    if p > alpha:
        print(f'Sample looks Gaussian(Fail to reject H0)')
    else:
        print(f'Sample does not look Gaussian(Reject H0)')

p_values = [p1, p2, p3]
for p in p_values:
    response(p)
    
"""
From visually inspection and the Shapiro-Wilk Test
none of the features are normally distributed.
"""
```

```python
"""
Take a look at BMI and Blood Pressure
"""
from sklearn.svm import LinearSVC
from sklearn.preprocessing import StandardScaler

features= pd.concat([df.BMI, df.BloodPressure], axis=1)
target = df.Outcome

# Standardize features
scaler = StandardScaler()
standardized_features = scaler.fit_transform(features)
color = ['black' if c == 0 else 'orange' for c in target]

svc = LinearSVC(C=1.0)
svc.fit(standardized_features, target) 

sns.set(rc={'figure.figsize':(10,10)})
plt.scatter(df.BMI, df.BloodPressure, c=color)

# Create the hyperplane
w = svc.coef_[0]
a = -w[0] / w[1]
xx = np.linspace(0, 55)
yy = a * xx - (svc.intercept_[0]) / w[1]

plt.plot(xx, yy)
```
