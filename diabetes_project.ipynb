```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

os.chdir('C:/Users/memdp/PYTHON_CODE/Datasets')
df = pd.read_csv('diabetes.csv')
df
```

```python
sns.set(rc={'figure.figsize':(15,10)})
sns.histplot(x=df.Age, y=df.BMI)
```

```python
"""
The majority of participants
"""
count=0
for x in df.Age:
    if x < 30:
        count += 1

print(count)
"""
Saving a png file containing the histogram and
a vertical line running through the mean and median
of the AGE feature in the dataset.

This is of importance to determine the age category
where most applicants fall under. 

We can determine bias or underrepresentations in 
the dataset early on by looking
at this particular feature.
"""
sns.set(rc={'figure.figsize':(15,10)})
ax = sns.histplot(data=df, x=df.Age)
ax.set(xlabel='Age', ylabel='Number of individuals')

plt.title('Age groups based on number of individuals', fontsize=20)
plt.axvline(df.Age.mean(), color='k', linestyle='dashed', label='mean')
plt.axvline(df.Age.median(), color='r', linestyle='dashed', label='median')
plt.legend(prop={'size': 17})
plt.savefig('C:/Users/memdp/PYTHON_CODE/Figures/AGE_histogram.png')


print(f'Age median: {df.Age.median()}')
print(f'Age mean: {round(df.Age.mean(), 2)}')

# print(f'Age < 30 count: {}')
```

```python
sns.set(rc={'figure.figsize':(15,10)})
ax = sns.histplot(data=df, x=df.Pregnancies, bins=17)
ax.set(xlabel='Pregnancies', ylabel='# of individuals')

plt.title('Histogram of individual(s) with number of pregnancies', fontsize=20)
plt.axvline(df.Pregnancies.mean(), color='k', linestyle='dashed', label='mean')
plt.axvline(df.Pregnancies.median(), color='r', linestyle='dashed', label='median')
plt.legend(prop={'size': 17})
plt.savefig('C:/Users/memdp/PYTHON_CODE/Figures/Pregnancies_histogram.png')
```

```python
"""
BMI mean is roughly equal to the 
BMI median, since there are not
many outliers.

The average BMI is 32, which means most 
people can be categorized as obese. This
makes sense given the people who parti-
cipated.
"""
sns.set(rc={'figure.figsize':(15,10)})
ax = sns.histplot(data=df, x=df.BMI)
ax.set(xlabel='BMI', ylabel='# of individuals')
plt.title('Histogram of individual(s) with BMI', fontsize=20)
plt.savefig('C:/Users/memdp/PYTHON_CODE/Figures/BMI_histogram.png')
print(f'BMI median: {df.BMI.median()}')
print(f'BMI mean: {round(df.BMI.mean(),2)}')
```
